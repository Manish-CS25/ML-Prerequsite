##solution 1.  To perform a matrix multiplication between a (n, k) and (k, m) matrix, it takes n*k*m multiplication and n*m*(k-1) operation.
               In resultant matrix , there is n*m elements , for each element we have to multilply k times and add k-1 time 


```python
## solution 2.

import time
# Timing using List of Lists

matrix1 = np.random.uniform(1, 10, size=(10, 5))

matrix2 = matrix1.T
start_time_list = time.time()

# Perform matrix multiplication using list of lists


def multiply_matrices(list1, list2):
    result_matrix = []
    for i in range(len(list1)):
        row = []
        for j in range(len(list2[0])):
            element = sum(a * b for a, b in zip(list1[i], (row[j] for row in list2)))
            row.append(element)
        result_matrix.append(row)
    return result_matrix


result_matrix_list = multiply_matrices(matrix1, matrix2)
end_time_list = time.time()
time_list = end_time_list - start_time_list
print(result_matrix_list)
```

```python
## solution 2.

import time


# Timing using NumPy
start_time_np = time.time()

# Perform matrix multiplication using NumPy


matrix1 = np.random.uniform(1, 10, size=(10, 5))

matrix2 = matrix1.T

result = np.dot(matrix1, matrix2)
end_time_np = time.time()
time_np = end_time_np - start_time_np
print(result)
```

```python


# Print the timings
print(f"Time taken using List of Lists: {time_list} seconds")
print(f"Time taken using NumPy: {time_np} seconds")
```

##by above comparison we can say multiplication of two matrices using numpy is faster than using list because numpy use low level optimization in computation.

```python
##solution 7.

def hierarchy(data, indent=0, index=None):
    for i, (key, value) in enumerate(data.items(), start=1):
        if index is not None:
            print("  " * indent + f"{i % 10}. {key}")
        else:
            print("  " * indent + f"{i % 10}. {key}")

        if isinstance(value, dict):
            hierarchy(value, indent + 1, i % 10)
        elif isinstance(value, list):
            for item in value:
                hierarchy(item, indent + 1, i % 10)
        else:
            print("  " * (indent + 1) + f" {key}: {value}")


students_record = {
    2022: {
        'Branch 1': {
             'Rollno.': 1, 'Name': 'N', 'Marks': {'Maths': 100, 'English': 70},
            
        },
        'Branch 2': {
            
        }
    },
    2023: {
        'Branch 1': {},
        'Branch 2': {},
    },
    2024: {
        'Branch 1': {},
        'Branch 2': {},
    },
    2025: {
        'Branch 1': {},
        'Branch 2': {},
    }
}


hierarchy(students_record)


    
```

```python
## solution 10.
import pandas as pd

## matrix of 20X5
matrix= np.random.uniform(1, 2, size=(20, 5))
print(matrix)


```

```python
##  creating a dataframe out of this matrix
column_names = ['a', 'b', 'c', 'd', 'e']
set = pd.DataFrame(matrix, columns=column_names)
print(set)
```

```python
#column with the highest standard deviation
highest_std_col = set.std().idxmax()
print(highest_col)
```

```python
# row with the lowest mean
lowest_mean_row = set.mean(axis=1).idxmin()
print(lowest_mean_row)
```

```python
## solution 11.

#new column "f" which is the sum of columns "a", "b", "c", "d", "e"
set['f'] = set[['a', 'b', 'c', 'd', 'e']].sum(axis=1)
print(set)
```

```python

# new column g is GT8 if f > 8 else LT8
set['g'] = np.where(set['f'] < 8, 'LT8', 'GT8')
print(set)
```

```python
# no. of rows where g is LT8
len(set[set["g"]=="LT8"])
```

```python
# standard deviation of column f where g is LT8
LT8_set=(set[set["g"]=="LT8"])
std_f= LT8_set['f'].std()

print(std_f)
```

```python
# standard deviation of column f where g is GT8
LT8_set=(set[set["g"]=="GT8"])
std_f= LT8_set['f'].std()

print(std_f)
```
